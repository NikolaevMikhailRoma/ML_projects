# Задание
В данном задании необходимо сгенерировать 3 облака двумерных точек и протестировать работу разных алгоритмов кластеризации на этих данных.



## Libs

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import adjusted_rand_score
```

## Gen and show

Генерация точек и вывод их на график

```python
X, y1 = make_blobs(n_samples=100, centers=3, n_features=2)  # Генерируем 30 двумерных точек, разбитых на 5 кластеров (X - координаты точек, y - метки принадлежности к кластерам)
print(X.shape, y.shape)                                   # Выводим размерность получившегося массива

fig, ax = plt.subplots(2, 3, figsize = (20,15))

ax[0,0].scatter(X[:, 0], X[:, 1], c=y)                        # Выводим точки на график с разбивкой точек по цвету
ax[0,0].set_xlabel('первоначальное распределение')

model = KMeans(n_clusters=3)
model.fit(X)
y = model.predict(X)
q = adjusted_rand_score(y, y1)
ax[0,1].scatter(X[:, 0], X[:, 1], c=y)                        # Выводим точки на график с разбивкой точек по цвету
ax[0,1].set_xlabel('KMeans с делением на 3 кластера, качество: '+str(round(q,2)))



model = KMeans(n_clusters=5)
model.fit(X)
y = model.predict(X)
q = adjusted_rand_score(y, y1)

ax[0,2].scatter(X[:, 0], X[:, 1], c=y)                        # Выводим точки на график с разбивкой точек по цвету
ax[0,2].set_xlabel('KMeans с делением на 5 кластеров качество: '+str(round(q,2)))

model = AgglomerativeClustering(n_clusters=3)
# model.fit(X)
y = model.fit_predict(X)
q = adjusted_rand_score(y, y1)

ax[1,0].scatter(X[:, 0], X[:, 1], c=y)                        # Выводим точки на график с разбивкой точек по цвету
ax[1,0].set_xlabel('AgglomerativeClustering с делением на 3 , качество: '+str(round(q,2)))

model = AgglomerativeClustering(n_clusters=5)
# model.fit(X)
y = model.fit_predict(X)
q = adjusted_rand_score(y, y1)

ax[1,1].scatter(X[:, 0], X[:, 1], c=y)                        # Выводим точки на график с разбивкой точек по цвету
ax[1,1].set_xlabel('AgglomerativeClustering с делением на 5 кластеров, качество: '+str(round(q,2)))

model = DBSCAN()
# model.fit(X)
y = model.fit_predict(X)
q = adjusted_rand_score(y, y1)

ax[1,2].scatter(X[:, 0], X[:, 1], c=y)                        # Выводим точки на график с разбивкой точек по цвету
ax[1,2].set_xlabel('DBSCAN, качество: '+str(round(q,2)))
plt.show()
```
